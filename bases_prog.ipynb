Regex avec les multiplicateurs

```python
# regex-3

import re
txt = "YabbbYaaYbbY"
print(f"txt={txt}")

regex_1 = re.compile("a*b+")
res_1= regex_1.findall(txt)
print(f"res_1={res_1}")

regex_2 = re.compile("Ya?b")
res_2= regex_2.findall(txt)
print(f"res_2={res_2}")


```

Regex avec un nombre d’occurrences limité

```python
# regex-4.py

import re
txt = "abaabbaaabbb"
print(f"txt={txt}")

regex_1 = re.compile("a{3}b{3}")
res_1= regex_1.findall(txt)
print(f"res_1={res_1}")

regex_2 = re.compile("a{2,3}b{2,3}")
res_2= regex_2.findall(txt)
print(f"res_2={res_2}")
```

Regex avec groupage des résultats

```python
import re

txt = "x=48.85 et y=2.35"

# Fix: Changed regex to correctly capture numbers separated by a dot
# and used a raw string for better regex handling.
regex = re.compile(r"([0-9]+)\.([0-9]+)")
res = regex.findall(txt)

print(txt)
for n1,n2 in res:
   print(f"n1={n1}; n2={n2}")
```

f. Taille des motifs

```python
# regex-6.py

import re

def apply_regex(regex, txt):
    print(f"\nApplication de la regex \"{regex}\" sur \"{txt}\"")
    regex = re.compile(regex)
    res = regex.finditer(txt)
    for e in res:
        print(f"e.group() = {e.group()}, e.start()={e.start()}, e.end()={e.end()}")


txt = "abcabcabcabcabc"
regex_1 = "(abc){2,5}"
apply_regex(regex_1, txt)
regex_2 = "(abc){2,5}?"
apply_regex(regex_2, txt)
```

Dans ce nouvel exemple, nous allons tenter d’extraire toutes les sous-chaînes de caractères entourées par des crochets. Pour cela, nous allons utiliser le code suivant :
regex-7.py

```python
import re

txt = "abc...[AAA], [1980],[Descartes] et [Covid-2 : 2020]...xyz"
print(f"txt={txt}")

# Fix: Changed regex to correctly extract content within square brackets
# using a raw string to avoid SyntaxError and to correctly parse regex.
regex = re.compile(r"\[([^\]]+)\]")
res = regex.findall(txt)
print(f"res={res}")
```

c. Gérer plusieurs exceptions
Exception-3.py

```python
try:
  a=int(input("Veuillez saisir un nombre : "))
  print(f"10/{a}={10/a}")
except ValueError:
  print("Vous avez saisi une valeur non numérique !")
except ZeroDivisionError :
  print("La division par zéro est impossible !")

print("Je suis la dernière instruction de ce programme")
```

d. Utiliser la clause finally
Exception-4.py

```python
 try:
  a=int(input("Veuillez saisir un nombre : "))
  print(f"10/{a}={10/a}")
except ValueError:
  print("Vous avez saisi une valeur non numérique !")
except ZeroDivisionError :
  print("La division par zéro est impossible !")
finally :
  print("Je suis finally !")

print("Je suis la dernière instruction de ce programme")
```

Le type booléen


```python
a, b, c = True, False, True

print(f"a = {a}, b = {b}, c = {c}")
print(f"Le type de la variable 'a' est : {type(a)}")
print(f"a = b : {a == b}")
print(f"a = c : {a == c}")

d = a+c
print(f"d = a+c : {d}")
print(f"Le type de la variable 'd' est : {type(d)}")
```

Les ensembles

```python
set_1 = {True,False,1,2,2,'SET'}
print(f"Le type de set_1 est : {type(set_1)}")
print(f"set_1 = {set_1}")

set_2 = {(1,2),(1,2),(1,2,3)}
print(f"set_2 = {set_2}")

set_3 = set([6,7])
print(f"set_3 = {set_3}")

set_3.update(set_1, {8,9})
print(f"set_3 = {set_3}")

set_3.discard(8)
set_3.remove(9)
print(f"set_3 = {set_3}")

```

Les conditions avec la structure if


```python
user_input = input("Veuillez saisir un nombre entier :")
n = int(user_input)

if n%2 == 0:
    print(f"{n} est un nombre pair")

if n%2 != 0:
    print(f"{n} est un nombre impair")
print("Fin du programme !")
```

Les conditions avec la structure if-elif-else


```python
user_input = input("Veuillez saisir un nombre entier :")
n = int(user_input)

if (n%2) == 0:
    print(f"{n} est un nombre pair")
    print(f"{n}**2 = {n**2}")

elif (n%3) == 0:
    print(f"{n} est un multiple de 3")
    print(f"{n}**3 = {n**3}")

elif (n%5) == 0:
    print(f"{n} est un multiple de 5")
    print(f"{n}**5 = {n**5}")

else:
    print(f"{n} est un nombre impair, mais il n'est pas un multiple de 3" \
    " ni un multiple de 5.")

print("Fin du programme !")
```

```python
def http_status(status):
    match status:
        case 200:
            return "OK"
        case 404:
            return "Not Found"
        case _:
            return "Unknown Status"
print(http_status(200))
```

```python
for value in [10,11,15,18]:
    print(f"value = {value}")
print("Fin du programme !")
```

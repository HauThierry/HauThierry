Encapsulation
Empêcher l'accès aux données d'un objet par un autre moyen que ceux définis.

```python
# Définir un attribut privé : _ATTRIBUT
class Animal:

  def __init__(self,race):
    self.race = race

simba_le_lion = Animal('Lion') # Lion est une instance de ma classe Animal

print(simba_le_lion.race)
```

```python
from os import replace
## Bonne pratique : définir des setteurs : set_attribut(self.attribut)

class Animal:

  def __init__(self,race):
    self.__race = race

  def set_race(self):
    return self.__race


simba_le_lion = Animal('Lion') # Lion est une instance de ma classe Animal

print(simba_le_lion.set_race())

simba_le_lion.__race = 'Lion de la savane'

print(simba_le_lion.set_race()) # On ne peut pas modifier encore l'attribut privé
```

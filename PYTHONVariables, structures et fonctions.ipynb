Les structures de contrôle


```python
#if else / elif permet de mettre des conditions
variable1000 = 'PSG'
if variable1000 == 'En Avant Guingamp !' :
  print("C'est l'EAG")
elif variable1000 == 'PSG' :
  print("C'est le Paris Saint Germain")
else:
  print("Ce n'est ni l'EAG, ni le PSG, c'est une autre équipe")
```

```python
#Boucles for <variable> in <list>
#NEVER REPEAT YOURSELF
variable1000 = 'PSG'
for i in [0,1,2,3]: #index des listes commence à 0
  print(i)
  print(variable1000)
for i in range(11):
  print(i)
```

```python
#while tant que <condition> pas résolue, on boucle.
i=0
while(i!=10):
  print(i)
  i = i+1

```

Les fonctions


```python
#Fonction def NOM FONCTION(ARGUMENT) Argument = les variables qui vont être utilisées dans cette fonction
def mean(number1, number2):
  mean_of_two_numbers = (number1 + number2)/2
  return mean_of_two_numbers

#print

mean(10,20)
```
